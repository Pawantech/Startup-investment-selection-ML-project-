```python
## Making optimal model using backward elimination
```

```python
import statsmodels.formula.api as sn
```

```python
#### MLR(MultiLinear Regression)####
#### MLR, y= b0 + b1*x1 + b2*x2 +.....+ bn*xn,  here b0 is constant which is not present in our data. ####
#### we can add b0 to dataset only if it is b0*x0 when x0=1 , therefore we need to add a column of 1 to dataset ####
#### here axis(parameter) determines if you want to add a line or column ####
#### axis=0(means adding line) and if axis=1(adding column) ####
```

```python
X = np.append(arr=np.ones((50,1)).astype(int), values=X, axis=1)
```

```python
X        # column with 1s are added
```

```python
#### Backward elimination process ####
```

```python
## making optimal model including all columns/independent vriable
## ordinary least square- method to make a model
X_opt = X[:,[0,1,2,3,4,5]]
regressor_OLS= sn.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()
```
